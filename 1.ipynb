### my answer 

> I tried to show the change in cnt over time.

```r
# your answer here
#here is the plot
ggplot(bike_df, 
       aes(x = instant, y = cnt, color = season)) +
  geom_point()+   
  geom_smooth()+
  xlab("Instant from 2011 to 2012") +
  ylab("count of total rental bikes") +
  ggtitle("Annual count of total rental bikes change, 2011 - 2012")
```
